# Global Ads Performance Analysis

### Objective
The objective of this project is to analyze global advertising performance data
for understand platform effectiveness, ad spend efficiency, and revenue trends.

This analysis was performed entirely using Python in Google Colab.

## Dataset Description

- Dataset - Global Ads Performance Dataset
- Data Type - Public dataset
- Records - Advertising performance across platforms and countries
- Key Metrics -
  - Impressions
  - Clicks
  - CTR (Click Through Rate)
  - CPC (Cost Per Click)
  - CPA (Cost Per Acquisition)
  - Ad Spend
  - Revenue
  - ROAS (Return on Ad Spend)

## Data Loading
The dataset was imported using pandas for analysis

```python

import pandas as pd

df = pd.read_csv("global_ads_performance_dataset.csv")

df.head()
```

## Data Quality Check
Basic checks were performed to ensure data quality before analysis

```python

df.info()

df.isnull().sum()

df.duplicated().sum()
```

### Observations
- No missing values were found in the dataset.
- No duplicate records were detected.
- The dataset is clean and suitable for analysis.

```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure()
plt.scatter(df['ad_spend'], df['revenue'])
plt.xlabel('Ad Spend')
plt.ylabel('Revenue')
plt.title('Ad Spend vs Revenue')
plt.show()
```

**Insight:**  
Higher ad spend generally leads to higher revenue, indicating a positive relationship
between investment and returns.

```python

df.groupby('platform')['revenue'].sum().plot(kind='bar')
plt.title('Revenue by platform')
plt.ylabel('Total Revenue')
plt.show()
```

**Insight:**  
Some platforms contribute significantly more revenue, suggesting better
performance or audience targeting.

```python

df.groupby('platform')['CTR'].mean().plot(kind='bar')
plt.title('Average CTR by Platform')
plt.ylabel('CTR')
plt.show()
```

**insight:**

The average CTR varies across advertising platforms, indicating differences in
user engagement and ad relevance. Platforms with higher CTR show stronger
interaction with audiences, while lower CTR platforms may require optimization
in targeting or creative strategy.

## Data Preparation
The Date column was converted to datetime format to enable time-based analysis.

```python
df['date'] = pd.to_datetime(df['date'])
```

```python

df.groupby('date')['revenue'].sum().plot()
plt.title('Revenue Over Time')
plt.ylabel('Revenue')
plt.xlabel('Date')
plt.show()
```

**insight:**

Revenue over time shows noticeable fluctuations, suggesting the influence of
campaign performance, budget changes, or seasonal factors. Time-based trends help
identify high-performing periods and support better planning for future
advertising strategies.

## Key Findings
- The dataset is clean with no missing or duplicate values.
- Ad spend and revenue show a positive correlation.
- Platform performance varies significantly across metrics like CTR and ROAS.
- Certain platforms deliver better return on ad spend and cost efficiency.
